```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

cap = cv2.VideoCapture("/dev/video2")
cap.set(cv2.CAP_PROP_CONVERT_RGB, 0)
#cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
#cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
```

```python
ret, buffer = cap.read()

buffer = buffer.reshape(480, 640, 2)
left, right = cv2.split(buffer)
```

```python
left_ = cv2.cvtColor(left, cv2.COLOR_BAYER_GB2GRAY)
right_ = cv2.cvtColor(right, cv2.COLOR_BAYER_GB2GRAY)

plt.imshow(cv2.hconcat([left_, right_]))
plt.show()
```

```python
stereo = cv2.StereoBM_create(numDisparities=16*1, blockSize=15)
disparity = stereo.compute(left_, right_)

plt.imshow(disparity)
plt.show()
```
